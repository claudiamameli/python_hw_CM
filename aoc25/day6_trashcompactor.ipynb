Additiong and multiplication, follow columns instead of the row and use the last row as either + o *.

```python
hw = [['123', '328',  '51', '64'], 
 ['45', '64',  '387', '23'], 
  ['6', '98',  '215', '314'],
['*',   '+',   '*',   '+']] 
```

```python
symbol = '*'

curr = ('6' + symbol + '45' + symbol + '123')

eval(curr) # I can use eval to change the string into a mathematical formula and calculare the final result!
```

```python
def create_formulas(hw):
    formulas = [''] * len(hw[0])
    
    for col in range(len(hw[0])):
        curr = ''
        symbol = None
        for row in hw[::-1]:
            if row[col].isnumeric() == False:
                symb = row[col]
                continue
            if curr == '':
                curr = row[col]
            else: 
                curr = curr + symb + row[col]
        formulas[col] = curr
    return(formulas)
```

```python
formulas = create_formulas(hw)
formulas
```

```python
total = 0 
for f in formulas:
    total += eval(f)   
total
```

```python
hw = [['123', '328', ' 51', '64 '],
[' 45', '64 ', '387', '23 '],
['  6', '98 ', '215', '314'],
['*  ', '+  ', '*  ', '+  ']]
```

```python
hw2 = list(zip(*hw))
```

```python
row = hw2[0]
row
```

```python
def read_columns_correctly(row):
    new_numbers = [''] * len(row)
    for num in row:
        symb = row[-1]
        curr_num = num
        if num.strip().isnumeric():
            for c in range(len(curr_num)):
                new_numbers[c] += curr_num[c]
        else:
            new_numbers[-1] = curr_num
    return(new_numbers)
```

```python
new_lists = []
for row in hw2:
    new_lists.append(read_columns_correctly(row))
```

```python
new_lists
```

```python
def create_formulas_pro(new_lists):
    formulas = []
    for row in new_lists:
        curr = ''
        for i in row[::-1]:
            if i.strip().isnumeric() == False:
                symb = i
                continue
            if curr == '':
                curr = i
            else:
                curr = curr + symb + i
        formulas.append(curr)
    return(formulas)
```

```python
new_formulas = create_formulas_pro(new_lists)
```

```python
new_formulas
```

```python
total = 0 
for f in new_formulas:
    total += eval(f)   
total
```
