````python
def find_top_links(coords):
    distances = []
    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            distances.append([math.dist(coords[i], coords[j]), 
                              coords[i], 
                              coords[j]])
    links = []
    distances.sort()
    for link in distances:
        links.append(link[1:])
    return(links)
````

````python
links = find_top_links(coords)
````

````python
circuits = [links[0]]
counter = 0
while counter <= 10:
    for link in links[1:]:
        a = link[0]
        b = link[1]
        for i in range(len(circuits)):
            if a in circuits[i] or b in circuits[i]:
                #print('match found')
                if a in circuits[i] and b in circuits[i]:
                    #print('both boxes already in')
                    break
                if b not in circuits[i]:
                    #print('adding b')
                    circuits[i].append(b)
                    counter += 1
                    if counter > 10:
                        break
                if a not in circuits[i]:
                    #print('adding a')
                    circuits[i].append(a)
                    counter += 1
                    if counter > 10:
                        break
                break
        else:
            #print('creating new circuit')
            circuits.append([a,b])
````
