````python
def find_top_links(coords):
    distances = []
    for i in range(len(coords)):
        for j in range(i + 1, len(coords)):
            distances.append([math.dist(coords[i], coords[j]), i, j])
    links = []
    distances.sort()
    for link in distances:
        links.append(link[1:])
    return(links)
````

````python
def merge_circuits(circuits, i, j):
    new = min(circuits[i], circuits[j])
    to_change = max(circuits[i], circuits[j])
    for i, c in enumerate(circuits):
        if c == to_change:
            circuits[i] = new
    return(circuits)
````

````python
def find_circuits(coords, links, limit):
    circuits = [-1] * len(coords)
    counter = 0
    curr_circuit = 0
    for l in links:
        i, j = l
        if circuits[i] == -1 and circuits[j] == -1:
            circuits[i] = curr_circuit
            circuits[j] = curr_circuit
            curr_circuit += 1 
            counter += 1
        elif circuits[i] != -1 and circuits[j] == -1:
            circuits[j] = circuits[i]
            counter += 1
        elif circuits[i] == -1 and circuits[j] != -1:
            circuits[i] = circuits[j]
            counter += 1
        elif circuits[i] != -1 and circuits[j] != -1:
            circuits = merge_circuits(circuits, i, j)
            counter += 1
        if counter >= limit:
            break
    return(circuits)
````

````python
def final_solution(circuits):
    unique_circuits = set(circuits)
    
    top3 = [0, 0, 0]
    for c in list(unique_circuits):
        if c == -1:
            continue
        n = circuits.count(c)
        if top3[-1] < n:
            top3.pop()
            top3.append(n)
            top3.sort(reverse=True)
    final = 1
    for n in top3:
        final = final * n
    return(final)
````

<br/>This should be easy if I just loop through all the links and not just 1000, so I take away the limit and I can break the loop when all the boxes are in the same circuit, so if the lenght of the unique numbers of circuits is 1!

````python
def find_final_junction(coords, links):
    circuits = [-1] * len(coords)
    counter = 0
    curr_circuit = 0
    for l in links:
        i, j = l
        if circuits[i] == -1 and circuits[j] == -1:
            circuits[i] = curr_circuit
            circuits[j] = curr_circuit
            curr_circuit += 1 
            counter += 1
        elif circuits[i] != -1 and circuits[j] == -1:
            circuits[j] = circuits[i]
            counter += 1
        elif circuits[i] == -1 and circuits[j] != -1:
            circuits[i] = circuits[j]
            counter += 1
        elif circuits[i] != -1 and circuits[j] != -1:
            circuits = merge_circuits(circuits, i, j)
            counter += 1
        if (set(circuits) != -1 #make sure we are not breaking at the beginning
            and len(set(circuits)) == 1): # break when all are connected!
            break 
    return(circuits, i, j)
````
