```python
def check_repeat(num):
    check = set()
    response = []
    final = []
    n = len(num) // 2 + 1
    for hop in range(1, n):
        seq = num[0:hop]
        times = int(len(num) / len(seq))
        if num == seq * times:
            response.append('T')
        else:
            response.append('F')
        if 'T' in response:
            final.append("T")
        else:
            final.append("F")
    return(final)
```

```python
def find_invalid_ids(ids):
    total = 0 
    for ID in ids: 
        i = ID.find('-')
        a = int(ID[:i])
        b = int(ID[i+1:])
        for num in range(a,b+1):
            num = str(num)
            result = check_repeat(num)
            if 'T' in result:
                total += int(num)
    return(total)
```
