One of the younger Elves was playing on a gift shop computer and managed to add a whole bunch of invalid product IDs to their gift shop database!
- invalid IDs are made by sequences of digits repeated twice
- Solution is adding the valid IDs together. 

### Practice

```python
s = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
```

```python
ids = s.split(',')
ids
```

```python
total = 0 
for ID in ids: 
    i = ID.find('-')
    a = int(ID[:i])
    b = int(ID[i+1:])
    for num in range(a,b+1):
        num = str(num)
        if num[0:3] == num[3:6]:
            print(ID, num)
            total += int(num)
            continue
        elif num[0:2] == num[2:4]:
            print(ID, num)
            total += int(num) 
            continue
        elif num[0] == num[1]:
            print(ID, num)
            total += int(num) 
            continue
        else:
            continue

# This is taking too many numbers into consideration, perhaps if I just split the IDs in half I can check the two halves against each other?
```

```python
total = 0 
for ID in ids: 
    i = ID.find('-')
    a = int(ID[:i])
    b = int(ID[i+1:])
    for num in range(a,b+1):
        num = str(num)
        half = int(len(num) / 2 )
        half1 = num[:half]
        half2 = num[half:]
        if half1 == half2:
            total += int(num)
```

```python
print(total)
```
