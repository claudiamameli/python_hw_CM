````python
def joltage(batt):
    start = 0
    digits = []
    i = 0
    for j in reversed(range(0, 12)):
        end = len(batt) - j
        if len(batt[start:]) > j:
            digit, i = find_digit(batt, start, end)
            digits.append(digit)
            start = i + 1
        else:
            start = i + 1
            digits.extend(batt[start:])
            break
    return(''.join(digits))
````
