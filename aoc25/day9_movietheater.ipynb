# Day 9 - Movie theater
Elves are redecorating the theater by switching square tiles in the big grid they form. Some of the tiles are red `#`. 
- find the largest rectangle that uses red tiles for two of its opposite corners!
- we are given the coordinates of where the tiles are! so row and column.

```python
import math
```

```python
tiles = [[7,1],
[11,1],
[11,7],
[9,7],
[9,5],
[2,5],
[2,3],
[7,3]]
```

```python
def find_largest_areas(tiles):
    distances = []
    for i in range(len(tiles)):
        for j in range(i + 1, len(tiles)):
            distances.append([math.dist(tiles[i], tiles[j]), tiles[i], tiles[j]])
    
    distances.sort(reverse = True)
    
    diag = math.ceil(distances[0][0])
    area = round((diag ** 2) / 2)
    
    return area
```

```python
find_largest_areas(tiles)
```
