```python
tiles = [[7,1],
[11,1],
[11,7],
[9,7],
[9,5],
[2,5],
[2,3],
[7,3]]
```

```python
def find_largest_areas(tiles):
    distances = []
    for i in range(len(tiles)):
        for j in range(i + 1, len(tiles)):
            distances.append([math.dist(tiles[i], tiles[j]), tiles[i], tiles[j]])
    
    distances.sort(reverse = True)
    
    diag = math.ceil(distances[0][0])
    area = round((diag ** 2) / 2)
    
    return area
```

```python
def find_largest_new(tiles):
    final = 0
    for i in range(len(tiles)):
        first = tiles[i]
        for j in range(i+1, len(tiles)):
            second = tiles[j]
            area = (abs(first[0] - second[0]) + 1) * (abs(first[1] - second[1]) + 1)
            if area > final:
                final = area
    
    return(final)      
```
