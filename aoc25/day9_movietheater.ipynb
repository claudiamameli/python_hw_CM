````python
def construct_perimeter(tiles, red_tiles):
    perimeter = red_tiles.copy()
    for i in range(len(tiles)):
        if i == len(tiles) - 1:
            a = tiles[i]
            b = tiles[0]
        else:
            a = tiles[i]
            b = tiles[i + 1]
            #print(a,b)
        if i % 2 == 0: # horizontal
            r = a[1]
            c = [a[0], b[0]]
            # Sorting them will make my life easier when adding the new tiles!
            c.sort()
            #print(r, c)
            perimeter[r, c[0]+1:c[1]] = 'x'
        elif i % 2 != 0: # vertical 
            r = [a[1], b[1]]
            c = a[0]
            r.sort()
            #print(r, c)
            perimeter[r[0]+1:r[1], c] = 'x'
    return(perimeter)
````

````python
perimeter = construct_perimeter(tiles, red_tiles)
print(perimeter)
````

````python
def check_around(i, j, perimeter):
    col = perimeter[:, j]
    row = perimeter[i, :]

    if (np.any((col[:i] == 'x') | (col[:i] == '#')) and
        np.any((col[i+1:] == 'x') | (col[i+1:] == '#')) and
        np.any((row[:j] == 'x') | (row[:j] == '#')) and
        np.any((row[j+1:] == 'x') | (row[j+1:] == '#'))):
        return True
    else:
        return False
````

````python
def construct_floor(perimeter):
    floor = perimeter.copy()
    rows = len(floor)
    cols = len(floor[0])
    
    for i in range(rows):
        for j in range(cols):
            if check_around(i, j, perimeter) == True and floor[i, j] == '.':
                floor[i, j] = 'x' 
    return(floor)
````

````python
floor = construct_floor(perimeter)
````

````python
def find_largest_final(tiles, floor):
    final = 0
    for i in range(len(tiles)):
        first = tiles[i]
        for j in range(i+1, len(tiles)):
            second = tiles[j]
            
            r1 = min(first[1], second[1])
            r2 = max(first[1], second[1])
            c1 = min(first[0], second[0])
            c2 = max(first[0], second[0])
            
            if '.' not in floor[r1:r2+1, c1:c2+1]:
                area = (abs(first[0] - second[0]) + 1) * (abs(first[1] - second[1]) + 1)
                if area > final:
                    final = area

    return(final)    
````

````python
find_largest_final(tiles, floor)
````

````python
def solve_part2(tiles):
    red_tiles = create_original_floor(tiles)
    print('tiles created')
    perimeter = construct_perimeter(tiles, red_tiles)
    print('perimeter created')
    floor = construct_floor(perimeter)
    print('floor created')
    return(find_largest_final(tiles, floor))
````
