# Day 5 - Cafeteria

The cafeteria needs to organise their inventory to be able to figure out which ingredients are `fresh` and which are `spoiled`. <br/>
The ingredients are organised by a `number id`. If the `id` is within **ANY** `id range`, then it is fresh. 
- So you calculate HOW MANY of the ingredients are FRESH!

````python
ranges = ['3-5','10-14','16-20','12-18']
ids  = [1, 5, 8, 11, 17, 32]
````

````python
def total_ranges(ranges):
    fresh_id = set()
    for seq in ranges:
        seq = seq.split("-")
        a = int(seq[0])
        b = int(seq[1]) + 1
        for num in range(a,b):
            fresh_id.add(num) 
    return(fresh_id)
````
