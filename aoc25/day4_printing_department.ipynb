````python
def create_grid(prelim_grid):
    for row in range(len(prelim_grid)):
        prelim_grid[row] = "." + prelim_grid[row] + "."
    spacers = str("." * len(prelim_grid[0]))
    grid = [spacers]
    grid.extend(prelim_grid)
    grid.append(spacers)
    return(grid)
````

````python
def count_rolls_in_row(grid, row_n):
    total_rolls = 0
    prev = grid[row_n - 1]
    curr = grid[row_n]
    succ = grid[row_n + 1]
    for i in range(len(curr)): 
        if curr[i] == '@':
            total_adj = 0 
            total_adj += prev[i-1:i+2].count("@")
            total_adj += curr[i-1].count("@") + curr[i+1].count("@")
            total_adj += succ[i-1:i+2].count("@")
            if total_adj < 4:
                total_rolls += 1
    return(total_rolls)
````

````python
def count_and_remove(grid, row_n):
    total_rolls = 0
    prev = grid[row_n - 1]
    curr = grid[row_n]
    succ = grid[row_n + 1]
    new_curr = ''
    for i in range(len(curr)):
        if curr[i] == '@':
            total_adj = 0 
            total_adj += prev[i-1:i+2].count("@")
            total_adj += curr[i-1].count("@") + curr[i+1].count("@")
            total_adj += succ[i-1:i+2].count("@")
            if total_adj < 4:
                total_rolls += 1
                new_curr += "."
            else:
                new_curr += curr[i]
        else:
            new_curr += curr[i] 
    
    return(total_rolls, new_curr)
````

````python
total = 0 
new_grid = grid[:]
for row_n in range(1, len(grid[1:len(grid)])):
    final_rolls, new_curr = count_and_remove(grid, row_n)
    total += final_rolls
    new_grid[row_n] = new_curr
print(total)
````

````python
def maximise_roll_extraction(grid):
    total = 0
    current_rolls = 1 # is there a better way to say this?
    while current_rolls > 0:
        current_rolls = 0
        new_grid = grid[:]
    
        for row_n in range(1, len(grid[1:len(grid)])):
            final_rolls, new_curr = count_and_remove(grid, row_n)
            current_rolls += final_rolls
            new_grid[row_n] = new_curr
    
        if current_rolls == 0:
            break
    
        total += current_rolls
        grid = new_grid[:]
    return(total)
````
