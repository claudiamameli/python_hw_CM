# Day 7 - Laboratories
The teleporter is leaking agic smoke! There are a lot of spare parts, manuals and diagnostic equipements. we Get error code `0H-N0`. This means thre is an issue with one of the tachyon manifold. <br/>

- Tachyon beams enters the manifold at location marked `S` and the beams always move **downwards**.
- beams pass freely through empty spaces `.`
- if beam encounters a splitter `^`, the beam is stopped and a new beam continues from the immediate left and immediate right of the splitter.
- AIM: How many times is a beam split?

```python
filepath = 'day7_trial.txt'
```

```python
tach = []
with open(filepath, 'r') as infile:
    for line in infile:
        line = line.rstrip('\n')
        split = list(line)
        tach.append(split)
```

```python
def check_start_split(tach, line_n, c, counter):
    if line_n + 1 >= len(tach):
        return tach, counter
        
    i = c
    if tach[line_n+1][c] == '^':
        counter += 1
        tach[line_n+1][i+1] = '|'
        tach[line_n+1][i-1] = '|'
    return(tach, counter)
```

```python
def count_splits(tach):
    counter = 0
    for line_n in range(len(tach)):
        line = tach[line_n]
        if 'S' in line:
            i = line.index('S')
            tach[line_n+1][i] = '|'
        for c in range(len(line)):
            if line[c] == '|':
                tach, counter = check_start_split(tach, line_n, c, counter)
            if line_n + 1 < len(tach) and line[c] == '|'  and tach[line_n+1][c] == '.':
                tach[line_n+1][c] = '|'
    return(counter)
```

```python
count_splits(tach)
```
