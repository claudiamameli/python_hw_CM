### Part 1 :) 

We need to find the password!

- Given: a sequence of rotations, 1 per line. 
- Return: number of times the dial is at '0'.

```python
filepath = 'day1_trial.txt'
directions = []
with open (filepath, 'r') as infile:
    for line in infile.readlines():
        cleaned = line.rstrip()
        directions.append(cleaned)
```

```python
def check_position(dial):
    if dial >= 100:
        dial -= 100
    elif dial < 0:
        dial += 100
    else:
        dial = dial
    return(dial)
```

```python
prelim = []
dial = 50
for i in range(len(directions)):
    steps = 0
    if len(directions[i]) > 3:
        steps = directions[i][-2:]
    else:
        steps = directions[i][1:]
    if directions[i][0] == 'L':
        dial -= int(steps)
        dial = check_position(dial)
        prelim.append(dial)
    elif directions[i][0] == 'R':
        dial += int(steps)
        dial = check_position(dial)
        prelim.append(dial)

prelim.count(0)
```
