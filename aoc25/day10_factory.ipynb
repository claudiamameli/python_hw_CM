# Day 10 - Factory
Factory machine are all offline - There are one indicator light diagrams, in the manual with button wiring schemstics and joltage requirements for each machine.

```python
light = ['.', '.', '.', '.',]
manual = ['.','#', '#', '.']
wiring_scheme = [[3], [1,3], [2], [2,3], [0,2], [0,1]]
# joltage = {3,5,4,7}
```

```python
def switch(final_position, wires):
    for button in wires:
        if final_position[button] == '.':
            final_position[button] = '#'
        elif final_position[button] == '#':
            final_position[button] = '.'
    return final_position
```

```python
counter = 0
final_position = light[:]
for wires in wiring_scheme:
    if final_position == manual:
        print(counter)
        break
    else: 
        counter += 1
        final_position = switch(final_position, wires)
```

```python
def check_1_switch(light, manual, wiring_scheme):
    counter = 0
    final_position = light[:]
    for wires in wiring_scheme:
        if final_position == manual and counter == 1:
            return(True)
            break
        else:
            counter = 0 
            final_position = light[:]
            counter += 1
            final_position = switch(final_position, wires)
    return(False)
```

```python
check_1_switch(light, manual, wiring_scheme)
```

```python
light = ['.', '.', '.', '.',]
manual = ['.','#', '#', '.']
wiring_scheme = [[3], [1,3], [2], [2,3], [0,2], [0,1]]
```

```python
def check_2_switch(light, manual, wiring_scheme):
    counter = 0
    final_position = light[:]
    for wire_a in wiring_scheme:
        for wire_b in wiring_scheme:
            if final_position == manual and counter == 2:
                return(True, final_position)
                break
            elif final_position != manual and counter == 2:
                counter = 0
                final_position = light[:]
            elif counter == 1:
                counter += 1
                final_position = switch(final_position, wire_b)
            else:
                counter += 1
                final_position = switch(final_position, wire_a)
    return(False, final_position)
```

```python
check_2_switch(light, manual, wiring_scheme)
```
