# Variable scope
https://galicae.github.io/python-novice/11-scope.html

The scope of a variable is the part of a program that can 'see' that variable.
The part of a program in which a variable is visible is called its *scope*.

````python
pressure = 103.9

def adjust(t):
    temperature = t * 1.43 / pressure
    return temperature
````

`pressure` is a ***GLOBAL VARIABLE***
- defined outside any particular function and visible anywhere.

`t` and `temperature` are ***LOCAL VARIABLES*** in the function.
- defined in the function
- not visible in the main program
NOTE: A fucntion patameter is a variable that is assigned a vaue when the function is called. 

````python
print('adjusted:', adjust(0.9))
print('temperature after call:', temperature)
````
