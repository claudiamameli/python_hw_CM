#### Division Types

```python
print('5 // 3:', 5 // 3)
print('5 / 3:', 5 / 3)
print('5 % 3:', 5 % 3)
```

```python
num_subjects =  599
num_per_survey = 42
```

```python
print('Needed number of surveys:', ((num_subjects -1 )//(num_per_survey + 1)))
```

The reason why we need a -1 and a +1 is because we might want to make the function as general as possible. Specifically, the -1 is used in case the numbers of surveys is perfectly fitted into the number of subjects, in that cse the +1 at the end might mess up out final result because this would create too many surveys. The +1 is used because we might want to round up so that no one is left out.

#### Complex numbers
Python provides complex numbers, which are written as 1.0+2.0j. If val is a complex number, its real and imaginary parts can be accessed using dot notation as val.real and val.imag.

```python
a_complex_number = 6 + 2j
print(a_complex_number.real)
print(a_complex_number.imag)
```

To create a complex number you might use the complex() function.

```python
x = 5 
x = complex(5)
print(x)
```
