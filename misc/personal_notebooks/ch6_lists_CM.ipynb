#### Appending

```python
primes = [2, 3, 5]
print("primes is initially:", primes)
primes.append(7)
print("primes has become:", primes)
```

#### Extending 
This allows us to combine two lists

```python
teen_primes = [11, 13, 17, 19]
middle_aged_primes = [37, 41, 43, 47]
print('primes is currently:', primes)
primes.extend(teen_primes)
print('primes has now become:', primes)
primes.append(middle_aged_primes) #in this case the last elements will be a sub-list
print('primes has finally become:', primes)
```

#### Delition
_del list_name[index]_ &rarr; allows us to completely delete the item

```python
primes = [2, 3, 5, 7, 9]
print('primes before removing last item:', primes)
del primes[4]
print('primes after removing last item:', primes)
```

#### Sorting is tricky, we have 2 types of sorting commands
**sorted()** &rarr; outputs a result and so we can actually assign it to a new variable <br/>
**_list_.sort()** &rarr; sorts the item within the list without outputting anything

```python
# Program A
letters = list('gold')
result = sorted(letters)
print('letters is', letters, 'and result is', result)
```

```python
# Program B
letters = list('gold')
result = letters.sort()
print('letters is', letters, 'and result is', result)
```

### Careful with assigning!

```python
#Program A
old = list('gold')
new = old      # simple assignment
new[0] = 'D'
print('new is', new, 'and old is', old)
```

Here "new = old" makes a REFERENCE to the list "old"! So both get changed.

```python
# Program B
old = list('gold')
new = old[:]   # assigning a slice
new[0] = 'D'
print('new is', new, 'and old is', old)
```

Here instead, new = "old[:]" creats a NEW object, so we only change the new one.
