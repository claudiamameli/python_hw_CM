#### Trying to make sense of the ^ sign 

^ --> "exclusive OR" operator
Binary represents numbers only 0s and 1s --> computer language. 

0 → 0000 <br />
1 → 0001 <br />
2 → 0010 <br />
3 → 0011 <br />
4 → 0100 <br />
5 → 0101 <br />
6 → 0110 <br />
7 → 0111 <br />
8 → 1000 <br />
9 → 1001 <br />

Each digit(bit) represent a power of 2:
so 0101 = (0x8) + (1x4) + (0x2) + (1x1) = 5

1 --> bits are different 
0 --> bits are the same

0101 (5)
0011 (3)
________
0110 ----> 6 !


```python
1^1 # 0001 | 0001 = 0000 
```

```python
1^3 # 0001 | 0011 = 0010 
```

```python
1^5 # 0001 | 0101 = 0100 
```

```python
1^7 # 0001 | 0111 = 0110
```

```python
1^10 # 0001 | 1010 = 1011
```

```python
5^3 # 0101 | 0011 = 0110
```

#### Extra challenges - slicing

```python
species_name = 'Acacia buxifolia'
```

```python
print(species_name)
```

```python
species_name[2:8] #acia bu --> F, cuts before the character!
```

```python
species_name[11:] #error --> cuts before the 11th character and goes until end
```

```python
species_name[:4] #acaci --> cuts before the 4th character
```

```python
species_name[:] #full word --> T
```

```python
species_name[11:-3] #fo --> T
```

```python
species_name[-5:-3] #fo --> T, both start from the end
```

```python
species_name[0:20] #prints the all string --> T
```
