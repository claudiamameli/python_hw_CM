# Conditionals  - IF statements

Syntax is very similar to for loops! 
``` python
if <consdition>:
    <do something> 
else:
    <do something else>
```
Common conditionals are: <br/>
`x == 8` &rarr; x is 8<br/>
`x != 9` &rarr; x is not 8

````python
number = 4
if number > 0: 
    print("This is a positive number")
else:
    print("This is a negative number")
````

Here we might not be able to account for specific cases, like 0!

````python
number = 0
if number > 0: 
    print("This is a positive number!")
if number < 0:
    print("This is a negative number!")
if number == 0:
    print("This is 0!")
````

Although, this is a little convoluted and we might want to have proper functions for it.

````python
number = 0
if number > 0: 
    print("This is a positive number!")
elif number < 0: #used elif for more defined choices!
    print("This is a negative number!") 
else:
    print("This is 0!")
````

````python
fruits = ["apple", "grapes", "orange", "banana", "pomelo", "ananas", "kiwi"]

for fruit in fruits:
    if fruit == "banana":
        print("The fruit is a banana :)")
    else:
        print("The fruit is not a banana :(")
````

We might want to also look in which position the banana is!

````python
fruits = ["apple", "grapes", "orange", "banana", "pomelo", "ananas", "kiwi", "banana"]

for i in range(len(fruits)):
    # print(i, fruits[i])
    fruit = fruits[i]
    if fruit == "banana":
        print("The fruit in position", i, "is a banana :)")
 #  else:
 #      print("The fruit is not a banana :(")
````
