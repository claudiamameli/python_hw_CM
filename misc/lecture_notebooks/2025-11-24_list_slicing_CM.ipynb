# List bounds and slicing 
Revisiting some notes regarding the SUBS exercise

```python
s = "GATATATGCATATACTT"
```

```python
# The string does not care if we go out of bounds with slicing for a string 
s[15:19]
```

```python
# Although if we ask for a specific index in the list, this will not work 
s[19]
```

```python
# Similarly also with dictionaries we will get an error 
dic = {"foo" : "bar"}
dic[5]
```

```python
" ".join(s)
```

```python
rosalind_string = ""
for char in s:
    rosalind_string = rosalind_string + " " + char
print(rosalind_string.strip())
```

```python
print(*s)
```
