### Nested dictionaries
In this case we have a **key** which is the name of the student and its **value** is another dictionary which in itself has another **key** as the subject and a **value** which is the grade.

```python
grades = {'Juan': {'Maths': 2}, 
            'John': {'History': 1, 'Biology': 3},
            'Gianni': {'Biology': 1}}
```

```python
grades['Juan']['History'] = 2
grades['Juan']['Biology'] = 1

grades['John']['Maths'] = 1

grades['Gianni']['History'] = 4
grades['Gianni']['Maths'] = 3
```

```python
print(grades)
```

# Loops

```python
for number in [1, 2, 3, 4, 5]:
    print(number)
```

```python
for student in grades:
    print(student)
```

```python
for number in [123, 5, 38, 41, 50]:
    double = number * 2 
    print(number, double)
    print('next') #if we have a print inside the loop we will have it printed out at each step
print('loop finished') #here we print it at the end!
```

```python
shopping = ['bread', 'milk', 'cheese']
times = [1, 4, 10]
market = ['Oefferl', 'Hofer', 'Hofer']
```

```python
for i in range(3):
    print('Buy', shopping[i], times[i], 'times in', market[i])
```

```python
for i in range(len(shopping)): #when we do not know how many object we have we can use range(len(<list>))
    print('Buy', shopping[i], times[i], 'times in', market[i])
```

```python
for i, item, in enumerate(shopping): #this does the same thing as above
    print(i, item)
    print('Buy', shopping[i], times[i], 'times in', market[i])
```
