## Reading files
We will need to read a ot of files for Rosalind and AOC, here's an easy and quick way to do it - we construct a function!

````python
file_path = "/Users/claudiamameli/Desktop/uni/mec9_programming/python/python_hw_CM/rosalind/rosalind_ini5.txt"
````

````python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
        return lines
````

````python
line_list = read_file_into_lines(file_path)
line_list
````
