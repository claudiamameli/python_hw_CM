#### Claring things up for complex numbers
i = imaginary number &rarr; "j" in python language <br />
Stuff like this is convenient for root and negative numbers which cannot be defined. Complex numbers are used a lot in physics and are some sort of a built-in coordinate system for numbers. For example when you see e-4 you know that it is an exponential. But for i/j this is just an imaginary variable.

```python
test = 6 +2j
```

```python
test
```

```python
test.imag, test.real
```

```python
1 + 2j + 3 + 5j
```

```python
(1 + 2j) * (1j)
```

So basically we are usiing the letters as "a" "b" or "c" in an arithmetic formula. You can solve square or roots that you could not use before.
