### Practice
In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'.

The reverse complement of a DNA string ___s___ is the string ___s^c___ formed by reversing the symbols of ___s___, then taking the complement of each symbol (e.g., the reverse complement of "GTCA" is "TGAC").

Given: A DNA string ___s___ of length at most 1000 bp.
Return: The reverse complement ___s^c___ of ___s___.
.

```python
s = "AAAACCCGGT"
sc = ""
complement = {"A":"T", "T":"A", "C":"G", "G":"C"}
for nucl in s:
    sc = complement[nucl] + sc    
print(sc)    
```

### Using conditions!

```python
s = "AAAACCCGGT"
sc = ""

for nucl in s:
    if nucl == "A":
        sc ="T" + sc
    if nucl == "T":
        sc = "A" + sc
    if nucl == "C":
        sc = "G" + sc
    if nucl == "G":
        sc = "C" + sc
print(sc)
```

```python
s = "AAAACCCGGT"
sc = ""
complement = {"A":"T", "T":"A", "C":"G", "G":"C"}
for nucl in s[::-1]: #with reverse slicing
    sc = sc + complement[nucl]
print(sc)    
```

### Using string methods

```python
s = "AAAACCCGGT"
sc = s[::-1]
sc = sc.replace("A", "t")
sc = sc.replace("T", "a")
sc = sc.replace("C", "g")
sc = sc.replace("G", "c")
sc = sc.upper()
print(sc)
```

Can also be done all in the same line!

```python
s = "AAAACCCGGT"
sc = s[::-1].replace("A","t").replace("T", "a").replace("C", "g").replace("G", "c").upper()
print(sc)
```
