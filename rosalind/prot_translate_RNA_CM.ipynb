# Translating RNA into Protein
https://rosalind.info/problems/prot/

Given: an RNA string `s` that corrispong to a strand of mRNA <br/>
Return: The protein string encoded by `s`

0. codon &rarr; AA
1. find all the codons in the `s`
2. for every codon, find corresponding AA

&rarr; dictionary = codon : AA = "AUG" : "M"

NOTE: to not have the "Stop" show up in the final string it's better to delete "Stop" from the dictionary completely and have an empty string.

```python
codon_dict = {
'UUU': "F",      'CUU': 'L',      'AUU': 'I',      'GUU': 'V',
'UUC': "F",      'CUC': 'L',      'AUC': 'I',      'GUC': 'V',
'UUA': "L",      'CUA': 'L',      'AUA': 'I',      'GUA': 'V',
'UUG': "L",      'CUG': 'L',      'AUG': 'M',      'GUG': 'V',
'UCU': "S",      'CCU': 'P',      'ACU': 'T',      'GCU': 'A',
'UCC': "S",      'CCC': 'P',      'ACC': 'T',      'GCC': 'A',
'UCA': "S",      'CCA': 'P',      'ACA': 'T',      'GCA': 'A',
'UCG': "S",      'CCG': 'P',      'ACG': 'T',      'GCG': 'A',
'UAU': "Y",      'CAU': 'H',      'AAU': 'N',      'GAU': 'D',
'UAC': "Y",      'CAC': 'H',      'AAC': 'N',      'GAC': 'D',
'UAA': "",   'CAA': 'Q',      'AAA': 'K',      'GAA': 'E',
'UAG': "",   'CAG': 'Q',      'AAG': 'K',      'GAG': 'E',
'UGU': "C",      'CGU': 'R',      'AGU': 'S',      'GGU': 'G',
'UGC': "C",      'CGC': 'R',      'AGC': 'S',      'GGC': 'G',
'UGA': "",   'CGA': 'R',      'AGA': 'R',      'GGA': 'G',
'UGG': "W",      'CGG': 'R',      'AGG': 'R',      'GGG': 'G'}
```

```python
s = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
```

```python
counter = 0
aa = ""
for codon in s:
    triplet = s[counter:counter+3]
    if triplet == '':
        exit
    else:
        aa = aa + codon_dict[triplet] 
        counter = counter + 3
print(aa)
```
