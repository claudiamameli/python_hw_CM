### Practice
Given two positive integers ___a___
 and ___b___ (a<b<10000). <br/>
Return the ___sum of all odd integers___ from ___a___ through ___b___, inclusively.

```python
a = 100
b = 200 + 1 #this is important if we want the full range to be included
total = 0
```

```python
for num in range(a,b):
     if num % 2 != 0: #we check if the current number is an odd by dividing it by 2 and checking if the result is not equal to 0
         total = total + num
print(total)
```
