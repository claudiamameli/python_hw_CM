### Practice 
Given a string ___s___ of length at most 10000 letters. <br/>
Return the number of occurrences of ___each word___ in ___s___, where words are separated by spaces. Words are case-sensitive, and the lines in the output can be in any order.

```python
s = "We tried list and we tried dicts also we tried Zen"
```

```python
s = s.split(" ")
```

```python
print(s)
```

```python
word_count = {}
```

```python
for word in s:
    word_count[word] = s.count(word)
```

```python
print(word_count)
```

```python
for key, value in word_count.items():
    print(key, value)
```

### Solution in class

```python
words = ["and", "and", "test", "kiwi"]

frequencies = {}
for word in words:
    frequencies[word] = 0

for word in words:
    frequencies[word] = frequencies[word] + 1

print(frequencies)
```

```python
for key, value in frequencies.items():
    print(key, value)
```
