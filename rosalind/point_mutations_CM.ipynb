### Practice
Given two strings ___s___ and ___t___ of equal length, the Hamming distance between ___s___ and ___t___, denoted ___dH(s,t)___, is the number of corresponding symbols that differ in ___s___ and ___t___.

Given: Two DNA strings s and t of equal length (not exceeding 1 kbp).
Return: The Hamming distance dH(s,t).

```python
s = "GAGCCTACTAACGGGAT"
t = "CATCGTAATGACGGCCT"
mutations = 0 

for position in range(len(s)):
    if s[position] != t[position]:
        mutations = mutations + 1
print(mutations)
    
```
