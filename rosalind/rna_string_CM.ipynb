### Practice

Given a DNA string (t), its transcribed RNA string (u) is formed by replacing all occurrences of 'T' in (t)
 with 'U' in (u).



```python
dna = "GATGGAACTTGACTACGTAAATT"
```

```python
rna = dna.replace("T", "U")
```

```python
print(rna)
```

### Using loops and conditionals!

```python
dna = "GATGGAACTTGACTACGTAAATT"
```

```python
rna = ""
for nucl in dna:
    if nucl == "T":
        rna = rna + "U"
    else:
        rna = rna + nucl
print(rna)
```
