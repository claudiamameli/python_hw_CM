### Theory
Something to keep in mind is that whenever a fucntion for reading is used, the pointer stays in the same point in the text file, so we have to reset it if we want to start from the beginning. Everythime we run one spcific function it will continue from where we left off.

To access a file, you must first open it. To do so, you can use the ```open()``` function, which takes two parameters: the name of the target file and the mode. Three modes are available:

```r``` - read mode (the file is opened for reading) <br/>
```w``` - write mode (the file is opened for writing, and if a file having the same name exists, it will be erased) <br/>
```a``` - append mode (the file is opened for appending, which means that data is only to be added to the existing data in the file)<br/>



````python
f = open('file_ini5', 'r')
````

````python
f.seek(0)
f.read()
````

````python
f.seek(0)
f.read(10) #reads n number of characters
````

If you used one read function and then another, the pointer is at the end, you can reset it with ```<filename>.seek(0)```

````python
f.seek(0)
````

````python
line = f.readline().strip() #stip takes away the \n
line
````

````python
f.seek(0)
f.readlines() #here we separate each line using the \n
````

````python
f.seek(0)
f.readlines()[5] #look for specific lines!
````

````python
f.seek(0)
for line in f:
  print(line)
````

#### More examples of methods
```.split()``` &rarr; uses white spaces and \n as delimeters by default <br/>
```.splitlines()``` &rarr; returns a list of the lines in the string, breaking at line boundaries

````python
'Beautiful is better than ugly.\n'.split()
````

````python
'Explicit, is better, than implicit.'.split(",") #specify delimiter
````

````python
'Simple is\nbetter than\ncomplex.\n'.splitlines()
````

The command ```f.write(string)``` writes the contents of ```string``` to file ```f```. If you want to write something other than a string (an integer say), you must first convert it to a string by using the function ```str()```.

````python
f = open('output.txt', 'w')
inscription = ['Rosalind Elsie Franklin ', 1920, 1958]
s = str(inscription)
f.write(s)
````

````python
#can also be done in a loop!
for i in inscription:
  f.write(str(i) + '\n') #add \n at the end of each line!
````

### Practice 
Given a file containing at most 1000 lines. <br/>
Return a file containing all the even-numbered lines from the original file. Assume 1-based numbering of lines. 

````python
f = open("file_ini5", 'r')
````

````python
for line in f:
    print(line)
````

````python
f.seek(0)
line_num = 1
for line in f:
    if line_num % 2 == 0:
        print(line)
    line_num = line_num + 1
````
